### Matplotlib

Hay muchos paquetes en python que nos permiten visualizar datos, pero la madre de todos ellos es Matplotlib.

Este paquete se importa de la siguiente manera:

```python
import matplotlib.pyplot as plt     # Por convención se importa como "plt" 
```

Como primer ejemplo, analizaremos el tipo de mascotas que hay en los hogares de Medellín-Colombia.

Se tiene dos listas una con el porcentaje en forma decimal y otra con el tipo de mascota.

```python
cantidad = [0.70, 0.25, 0.05, 0.015, 0.015,]
animal = ["Perros", "Gatos", "Pajaros", "Peces", "Hamsters"]
```

Para trazar estos datos en una gráfica usamos la sentencia plt.plot con dos argumentos(X, Y): 

X -> Horizontal

Y -> Vertical

```python
plt.plot(animal, cantidad)
plt.show()
```

Otra manera de ver los datos es con el gráfico de dispersión, donde podemos observar de una forma más limpia los datos.

```python
plt.scatter(animal, cantidad)
plt.show()
```

### Nota:

En algunos casos si se tienen valores muy altos se recomienda poner los valores en una escala logarítmica. Así:

##### plt.xscale('log')
